```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_pickle('df_ohlcv.pkl')


df = df[df.index > pd.to_datetime('2020-03-01 00:00:00Z')]

(df['op']).rolling(1000).mean().plot(label='op')
(df['hi']).rolling(1000).mean().plot(label='hi')
(df['lo']).rolling(1000).mean().plot(label='lo')
(df['cl']).rolling(1000).mean().plot(label='cl')
plt.title('Transition of contract probability')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()


(df['hi']-df['cl'].shift(1)).rolling(1000).mean().plot(label='1')
(df['lo']-df['cl'].shift(1)).rolling(1000).mean().plot(label='2')
(df['hi']-df['lo']).rolling(1000).mean().plot(label='3')
plt.title('Transition of contract probability')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
```
